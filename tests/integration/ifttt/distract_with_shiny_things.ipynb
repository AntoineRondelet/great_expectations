```python
def plot_data(df, t):
    n = df.shape[0]
    df_sub = df.iloc[0:t]

    plt.plot(df.y, color="gray")
    plt.scatter(range(n), df.y, color="gray")

    plt.plot(df_sub.y, color="blue")
    plt.scatter(range(t), df_sub.y, color="blue", linewidth=2)

    plt.hlines([0], 0, n-1, color="lightgray")
    plt.hlines([-2,2], 0, n-1, linestyle="--")
    plt.hlines([-4,4], 0, n-1, linestyle="--", color="lightgray")
```

```python
def write_csv(df, t):
    df_sub = df.iloc[0:t]
    df_sub.to_csv("data/normal_data_over_time.csv", index=None)
```

```python
t = 0
plot_data(df, t)
```

```python
t += 1
plot_data(df, t)
write_csv(df, t)
```
